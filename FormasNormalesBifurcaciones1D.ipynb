```python
import scipy as sc
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline
```

```python
def nodoSilla(r,x):
    y=[]
    for i in x:
        y.append(r+i**2)
    return y
r=np.arange(-100,100,5)
x=np.arange(-10,10,0.5)
for j in r:
    plt.plot(x, nodoSilla(j,x))
```

```python
def transcritica(r,x):
    y=[]
    for i in x:
        y.append(r*i-i**2)
    return y
r=np.arange(-10,10,1)
x=np.arange(-20,20,0.5)
for j in r:
    plt.plot(x, transcritica(j,x))
```

```python
def tridente1(r,x):
    y=[]
    for i in x:
        y.append(r*i-i**3)
    return y
r=np.arange(-10,10,1)
x=np.arange(-2,2,0.1)
for j in r:
    plt.plot(x, tridente1(j,x))
```

```python
def tridente2(r,x):
    y=[]
    for i in x:
        y.append(r*i+i**3)
    return y
r=np.arange(-10,10,1)
x=np.arange(-2,2,0.1)
for j in r:
    plt.plot(x, tridente2(j,x))
```

```python
def bifurcacionEjercicio(r,x):
    y=[]
    for i in x:
        y.append(r*i+i**3-i**5)
    return y
r=np.arange(-5,5,1)
x=np.arange(-2,2,0.05)
for j in r:
    plt.plot(x, bifurcacionEjercicio(j,x))
```
