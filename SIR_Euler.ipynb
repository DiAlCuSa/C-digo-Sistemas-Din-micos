Dadas las ecuaciones

$$x_t=x(t)$$
$$y_t=x(t)$$
$$z_t=z(t)=1-x_t-y_t$$

que cumplen que para tiempos discretos 

$$x_{n+1}=x_n-h\alpha x_ny_n$$
$$y_{n+1}=y_n+h\alpha x_ny_n-h\beta y_n$$
$$z_{n+1}=1-x_{n+1}-y_{n+1}$$

Con las condiciones iniciales de:

$$x_0=0.5$$
$$y_0=0.5$$
$$z_0=1-x_0-y_0$$

```python
x0=0.5
y0=00.4
z0=1-x0-y0
h=0.001
n=10000
alpha=0.5  #Probabilidad de contagio a partir del contacto con el infectado
beta=0.1   #Probabilidad de recuperaci√≥n
x=[]
y=[]
z=[]
t=[]
x.append(x0)
y.append(y0)
z.append(z0)
t.append(0.0)
```

```python
for i in range(n):
    x.append(x[i]-h*alpha*x[i]*y[i])
    y.append(y[i]+h*alpha*x[i]*y[i]-h*beta*y[i])
    z.append(1-x[i]-y[i])
```

```python
plt.plot(x, label="x")
plt.plot(y, label="y")
plt.plot(z, label="z")
plt.xlabel("t")
plt.legend()
plt.show()
```
