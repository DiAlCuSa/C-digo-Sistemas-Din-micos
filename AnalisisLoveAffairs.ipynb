Tenemos un sistema dado por:
$$
R(t) = \frac{amor}{odio} \quad de\quad Romeo\quad por\quad Juliet\quad al\quad tiempo\quad t
$$

$$
J(t) = \frac{amor}{odio} \quad de\quad Julieta\quad por\quad Romeo\quad al\quad tiempo\quad t
$$
\begin{equation}
\frac{dR}{dt}=aR+bJ
\end{equation}

\begin{equation}
\frac{dJ}{dt}=cR+dJ
\end{equation}

```python
import scipy as sc
import numpy as np
import sympy as sy
import matplotlib.pylab as plt
import pylab as py
from scipy.optimize import fsolve
%matplotlib inline
from sympy import *
from numpy import *
```

```python
#ceroclinas
def cero_r(a,b,J):
    return -(b/a)*J

def  cero_j(c,d,J):
    return -(d/c)*J
```

```python

a1= 2.0
a2= -a1
b1= 3.0
b2= -b1
c1= 5.0
c2= -c1
d1= 6.0
d2= -d1
b0=0.0
c0=0.0

julieta = np.arange(0,20,0.1)
x , y  = np.meshgrid(np.arange(0,20,0.5), np.arange(-20,20,0.5)) 

cero_romeo_1=cero_r(a1,b1,julieta)
cero_julieta_1=cero_j(c1,d1,julieta)

cero_romeo_2=cero_r(a1,b2,julieta)
cero_julieta_2=cero_j(c2,d1,julieta)

cero_romeo_3=cero_r(a2,b1,julieta)
cero_julieta_3=cero_j(c1,d2,julieta)

cero_romeo_4=cero_r(a2,b2,julieta)
cero_julieta_4=cero_j(c2,d2,julieta)

cero_romeo_0=np.zeros(len(julieta))
cero_julieta_0=np.zeros(len(julieta))

u1 = a1*y+b1*x
v1 = c1*y+d1*x

u2 = a1*y+b2*x
v2 = c2*y+d1*x

u3 = a2*y+b1*x
v3 = c1*y+d2*x

u4 = a2*y+b2*x
v4 = c2*y+d2*x

u5 = a1*y+b0*x
v5 = c0*y+d1*x

u6 = a1*y+b0*x
v6 = c0*y+d2*x

lugarPuntoFijo1 = np.argwhere(np.diff(np.sign(cero_romeo_1 - cero_julieta_1)) != 0).reshape(-1) + 0
lugarPuntoFijo2 = np.argwhere(np.diff(np.sign(cero_romeo_2 - cero_julieta_2)) != 0).reshape(-1) + 0
lugarPuntoFijo3 = np.argwhere(np.diff(np.sign(cero_romeo_3 - cero_julieta_3)) != 0).reshape(-1) + 0
lugarPuntoFijo4 = np.argwhere(np.diff(np.sign(cero_romeo_4 - cero_julieta_4)) != 0).reshape(-1) + 0
lugarPuntoFijo0 = np.argwhere(np.diff(np.sign(cero_romeo_0 - cero_julieta_0)) != 0).reshape(-1) + 0
```

```python
#Matriz jacobiana simbólica
from sympy import Matrix
from sympy.abc import R, J,a,b,c,d

X = Matrix([a*R+b*J ,c*R+d*J ])
Y = Matrix([R, J])
J = X.jacobian(Y)
J
```

# Solución del sistema de ecuaciones

```python
# lado derecho de la ecuación 1 
def f(t,J,R,a,b):
    return a*R+b*J

#lado derecho de la ecuación 2
def g(t,J,R,c,d):
    return c*R+d*J
```

```python
def rk4(func1,func2,a,b,c,d,T,t0,n,y01,y02):
    
    t=np.zeros(n)
    t[0]=t0
    
    y1=np.zeros(n)
    y2=np.zeros(n)
    y1[0]=y01
    y2[0]=y02
    
    h=(T-t0)/n
    for i in range(0,n-1):
        t[i+1]=t0+i*h
        
        k1=func1(t[i],y1[i],y2[i],a,b)
        l1=func2(t[i],y1[i],y2[i],c,d)
        
        k2=func1(t[i]+h/2,y1[i]+(h/2)*k1,y2[i]+(h/2)*l1,a,b)
        l2=func2(t[i]+h/2,y1[i]+(h/2)*l1,y2[i]+(h/2)*l1,c,d)
        
        
        k3=func1(t[i]+h/2,y1[i]+(h/2)*k2,y2[i]+(h/2)*l2,a,b)
        l3=func2(t[i]+h/2,y1[i]+(h/2)*k2,y2[i]+(h/2)*l2,c,d)
        
        k4=func1(t[i]+h,y1[i]+h*k3,y2[i]+h*l3,a,b)
        l4=func2(t[i]+h,y1[i]+h*k3,y2[i]+h*l3,c,d)
        
        y1[i+1]=y1[i]+(h/6)*(k1+2*k2+2*k3+k4)
        y2[i+1]=y2[i]+(h/6)*(l1+2*l2+2*l3+l4)
        
        
    return t,y1,y2
```

```python
T=50
t0=0.0
n=1500
r0=2
j0=0
```

```python
t,r1,j1=rk4(f,g,a1,b1,c1,d1,T,t0,n,r0,j0)
t,r2,j2=rk4(f,g,a1,b2,c2,d1,T,t0,n,r0,j0)
t,r3,j3=rk4(f,g,a2,b1,c1,d2,T,t0,n,r0,j0)
t,r4,j4=rk4(f,g,a2,b2,c2,d2,T,t0,n,r0,j0)
t,r5,j5=rk4(f,g,a1,b0,c0,d1,T,t0,n,r0,j0)
t,r6,j6=rk4(f,g,a1,b0,c0,d2,T,t0,n,r0,j0)
```
