# EJEMPLO DE CAOS

\begin{equation}
    x_{t+1}=f(x)=4(1-x_t)x_t =4(x_t- x_t^2)
\end{equation}


Vamos a iterar con dos Condiciones Iniciales diferentes:

<ol>

  <li> $X_0$ = 0.523423 </li>
  <li> $X_0$ = 0.523424 </li>
  
</ol>

```python
import scipy as sc
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline
```

```python
#Definir las función
def iterar(t,x0, R):
    n=len(t)+1
    x = np.zeros(n)  
    x[0]=x0
    for i in range(0,n-1):
        x[i+1]=(x[i]-x[i]**2)*R
    return x
```

```python
#Definir los parámetros

x01= 0.523423
x02= 0.523424

T=sc.arange(0,100,1.0)
T1=sc.arange(0,101,1.0)
x1= iterar(T,x01,4.0)
x2= iterar(T,x02,4.0)
```

```python
print("t \t f(x) con x_01 \t f(x) con x_02")

for i in range(0,len(T)):
    tt=T[i]
    n1= x1[i]
    n2= x2[i]
    str0 = "%g \t %g \t%  g"%(tt, n1, n2)
    print(str0)
```

```python
#Gráfica de suseptibilidad a las condiciones iniciales de sistemas caóticos 
plt.plot(T1,x1, 'ro')
plt.plot(T1,x2, 'bs')
plt.xlabel("t")
plt.ylabel("x_t")
plt.title("Caos")

```

<ol>
<p>• Para $3.0000 < R < 3.4495$, hay un ciclo estable de periodo 2.</p>
<p>• Para $3.4495 < R < 3.5441$, hay un ciclo estable de periodo 4.</p>
<p>• Para $3.5441 < R < 3.5644$, hay un ciclo estable de periodo 8.</p>
<p>• Para $3.5644 < R < 3.5688$, hay un ciclo estable de periodo 16.</p>
<p>• Cuando $R$ se acerca a  $3.570$, hay ciclos estables de periodo $2n$, done el periodo de los ciclos incrementa cuando se acerca a $3.570$</p>
<p>• Para valores de $R > 3.570$, hay rangos soluciones periódicas y rangos de soluciones aperiódicas. </p>

</ol>

```python
#Iteraciones para diferentes rangos de "R"

X0= iterar(T,x01,2.0)
X1= iterar(T,x01,3.2)
X2= iterar(T,x01,3.5)
X3= iterar(T,x01,3.55)
X4= iterar(T,x01,3.566)
X5= iterar(T,x01,3.569)
X6= iterar(T,x01,3.571)
X7= iterar(T,x01,4.0)
```

```python
#Gráficas de diferentes periodicidades dependientes de los valores de "R"

plt.plot(T1,X1, 'bo', )  #periodo 2  "azul"
plt.plot(T1,X2, 'go')    #periodo 4  "verde"
plt.plot(T1,X3, 'co')    #periodo 8  "cyan"
plt.plot(T1,X4, 'mo')    #periodo 16  "magenta"
plt.plot(T1,X5, 'yo')    #periodo 2n  "amarillo"
plt.plot(T1,X6, 'ro')    #rangos estables e inestables "rojo"

plt.xlabel("t")
plt.ylabel("x_t")
plt.title("Periodicidades dependientes de R")
```

```python
#Gráfica de comparación de estado estable a caos

plt.plot(T1,X7, 'ro')
plt.plot(T1,X0, 'bo')
plt.xlabel("t")
plt.ylabel("x_t")
plt.title("Cambio de comportamiento")
```

# Definimos la función 

```python
def y(x,m): 
    n=len(x)
    y=np.zeros(n) 
    for i in range(0,n):
        if x[i]>=0.0 and x[i]<(1/m):
            y[i]=m*x[i]
        elif x[i]>=(1/m) and x[i]<1:
             y[i]=(m*x[i])-1.0
        else:
            print("Fuera de rango")
    return y
```

# Graficamos la función

```python
x=sc.arange(0.0,1.0,0.001)
y1=y(x,m)
plt.plot(x,y1)
plt.plot(x,x)#identidad
plt.xlabel("x_t")
plt.ylabel("x_{t+1}")
```

# Encontramos puntos fijos 

\begin{equation}
x^*=m x^* \quad con \quad m\neq0 \quad   \Longrightarrow \quad  x^*=0   \quad para \quad 0 < x_t < 1/m 
\end{equation}

\begin{equation}
x^*=m x^*-1 \quad con \quad m\neq0 \quad   \Longrightarrow \quad  x^*=\frac{1}{m-1}   \quad para \quad 1/m < x_t < 1
\end{equation}



# Encontramos la derivada de la función

\begin{equation}
\partial_t f(x)=m  \quad para \quad  0 < x < 1/m
\end{equation}

\begin{equation}
\partial_t f(x)=m \quad  para \quad  1/m < x_t < 1
\end{equation}

No depende de los puntos fijos, sólo depende de $m$. Como $1<m<2$ $\Longrightarrow$ creciente monónotona e inestable.

# Iteramos numéricamente



```python
#Definimos la función para iterar
def yit(x,m): 
    if x>=0.0 and x<(1/m):
        y=m*x
    elif x>=(1/m) and x<1:
         y=(m*x)-1.0
    else:
        print("Fuera de rango")
    return y

def iterar2(t,x0,m):
    n=len(t)+1
    x = np.zeros(n)  
    x[0]=x0
    for i in range(0,n-1):
        x[i+1]=yit(x[i],m)
    return x
```

```python
#Definimos los parámetros

equis0=0.2

tau=sc.arange(0,100,1.0)
tau1=sc.arange(0,101,1.0)
y1= iterar2(tau,equis0,m)
```

```python
print("t \t f(x)")

for i in range(0,len(tau)):
    tt=tau[i]
    N1= y1[i]
    str0 = "%g \t %g"%(tt, N1)
    print(str0)
```
