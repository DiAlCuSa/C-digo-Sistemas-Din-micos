```python
R=[0.5,0.6,0.8,0.9]
x0=0.9

x=[]
for j in range(len(R)):
    x1=[]
    x1.append(x0)
    for i in range(0,200):
        x1.append(R[j]*(1-x1[i])*x1[i])
    x.append(x1)  
print(x)
           
```

```python
for i in range(len(x)):
    plt.plot(x[i], label=i)
    plt.legend()
```
